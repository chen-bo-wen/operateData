### 修改后缀

```python
import os
import shutil

def remove_suffix(folder_path):
    # 遍历文件夹中的所有文件
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        print('file_path', file_path)
        # fileName, suffix = os.path.splitext(file)

        # 检查文件是否为图片
        if os.path.isfile(file_path) and any(filename.lower().endswith(ext) for ext in ['.jpg', '.jpeg', '.png']):

            # 检查文件名是否以"_EX"结尾
            if filename.endswith('.jpeg') :
                new_filename = filename.replace(".jpeg", ".jpg")
                # 构造新文件的路径
                new_file_path = os.path.join(folder_path, new_filename)
                print('new_file_path', new_file_path)

                # 移动并重命名文件
                shutil.move(file_path, new_file_path)
                print(f"已删除文件名后缀：{filename}")

            if filename.endswith('.png') :
                new_filename = filename.replace(".png", ".jpg")

                # 构造新文件的路径
                new_file_path = os.path.join(folder_path, new_filename)
                print('new_file_path', new_file_path)

                # 移动并重命名文件
                shutil.move(file_path, new_file_path)
                print(f"已删除文件名后缀：{filename}")

    print("处理完成")


if __name__ == "__main__":
    # 指定要检查的文件夹路径
    folder_path = '沙壳蛋'

    # 调用函数进行处理
    remove_suffix(folder_path)
```

```python
def rreplace(self, old, new, *max):
  count = len(self)
  if max and str(max[0]).isdigit():
    count = max[0]
  return new.join(self.rsplit(old, count))

 rreplace("str", "_", "-", 1) # 数字 1 就表示只替换一个
```

### <a id='111'>统一命名</a>

<a href='#top'>▲ Top</a>

os.rename(old_File_Path, New_File_Path)

```python
# walk_modifyName_combine
import os
import shutil

outputfile = open('output.txt', 'w')
outputfilename = open('outputname.txt', 'w')


def iter_files(rootDir):
    for root,dirs,files in os.walk(rootDir):
        for file in files:
            if any(file.lower().endswith(ext) for ext in ['.jpg', '.jpeg', '.png']): # 对于所有为图片的文件，不管是哪个层级的

                old_File_Path = os.path.join(root,file).replace('\\','/') # 旧图片路径

                oldFilePath_split = old_File_Path.split('/')
                # print(old_File_Path)

                Name_Folder = oldFilePath_split[-2] # 需要添加到图片的标签名称，如 AnBanDan、EAA 等文件夹名称

                Old_File_Name = oldFilePath_split[-1]
                Old_File_Name_split = Old_File_Name.split('_')
                Old_File_Name_split[0] = Name_Folder
                symbol = "_"
                
                New_File_Name = symbol.join(Old_File_Name_split)
                New_File_Path = os.path.join(root, New_File_Name).replace('\\','/')
                print(New_File_Path)
                outputfile.write(New_File_Name)
                outputfile.write('\r\n')

                os.rename(old_File_Path, New_File_Path)


root = os.getcwd()
rootDir = os.path.join(root, 'AB').replace('\\', '/')
iter_files(rootDir)
outputfile.close()
outputfilename.close()
```

### <a id='4'>遍历不同层级的同类型文件，统一命名并进行分类整合（os.walk）</a>

<a href='#top'>▲ Top</a>

shutil.copy(oldFilePath, newFolderPath) 是旧文件路径，到新文件夹路径 ！！！ 否则会报错 PermissionError: [Errno 13] Permission denied

新目录和旧目录没有放在同一个目录下！！! 不然遍历的时候也会遍历新目录。

```python
# walk_modifyName_combine
import os
import shutil

outputfile = open('output.txt', 'w')
outputfilename = open('outputname.txt', 'w')


def iter_files(rootDir):
    for root,dirs,files in os.walk(rootDir):
        for file in files:
            if any(file.lower().endswith(ext) for ext in ['.jpg', '.jpeg', '.png']): # 对于所有为图片的文件，不管是哪个层级的

                oldFilePath = os.path.join(root,file).replace('\\','/') # 旧图片路径

                oldFilePath_split = oldFilePath.split('/')
                # print(oldFilePath)
                outputfile.write(oldFilePath)
                outputfile.write('\r\n')

                EggFolder = oldFilePath_split[-3] 

                for ch in oldFilePath_split[-2]: 
                    if u'\u4e00' <= ch <= u'\u9fff':#(存在中文)
                        Eggindex = oldFilePath_split[-2].split('-') # 对于 如 “1-粉色-畸形” 形式 进行 split

                        if len(Eggindex) >= 3: # 存在有的情况 len 不大于 3，这时会报错
                            newFolder = Eggindex[2] # 新文件夹名称命名为：正常/畸形/。。。

                            jpgName = oldFilePath_split[-1] 
                            symbol = "_"
                            
                            newFolderPath = os.path.join('../蛋品数据（更新中）/新存储路径', newFolder)
                            
                            if not os.path.isdir(newFolderPath): # 如果目标不存在，则创建
                                os.makedirs(newFolderPath)
                                print(newFolderPath)

                            if EggFolder[0].isdigit(): # 开头如果是数字就放入图片新命名名称里，只取前 8 位数字，后面的中文如(496号开始错位)就不取
                                Eggindex_index = Eggindex[0]
                                seq = (EggFolder[0:8], Eggindex_index, jpgName)
                                finalName = symbol.join( seq )

                                # os.rename(file, os.path.join(newFolderPath, finalName).replace('\\','/'))
                                newfilepath = os.path.join(newFolderPath,finalName).replace('\\','/')

                                shutil.copy(oldFilePath, newfilepath)
                                print(newfilepath)
                                outputfilename.write(newfilepath)
                                outputfilename.write('\r\n')
                            else:
                                Eggindex_index = Eggindex[0]
                                seq = (Eggindex_index, jpgName)
                                finalName = symbol.join( seq )

                                newfilepath = os.path.join(newFolderPath,finalName).replace('\\','/')

                                # os.rename(file, os.path.join(newFolderPath, finalName).replace('\\','/'))
                                shutil.copy(oldFilePath, newFolderPath)
                                print(newfilepath)
                                outputfilename.write(newfilepath)
                                outputfilename.write('\r\n')

rootDir = '../蛋品数据（更新中）/新存储路径'
iter_files(rootDir)
outputfile.close()
outputfilename.close()
```

### <a id='5'>Same_Name_Folder_Combine 合并相同名称文件夹下的图片</a>

<a href='#top'>▲ Top</a>

```python
# Same_Name_Folder_Combine
# coding=utf-8
import os, random, shutil
import glob

outputfilename = open('outputname.txt', 'w')

def CombineFile():
    rootDir = os.getcwd()
    for root,dirs,files in os.walk(rootDir):
        for dir in dirs:
            oldFolderPath = os.path.join(root, dir).replace('\\','/')

            # 新文件路径不要放在和旧文件同一个目录下
            newFolderPath = os.path.join('../dataset', dir).replace('\\','/') # 存放要跳到上级目录才看到
            # print(newFolderPath)
            if not os.path.isdir(newFolderPath): # 如果目标不存在，则创建
                os.makedirs(newFolderPath)

            for file in os.listdir(oldFolderPath):
                if any(file.lower().endswith(ext) for ext in ['.jpg', '.jpeg', '.png']): # 对于所有为图片的文件，不管是哪个层级的
                    oldFilePath = os.path.join(oldFolderPath, file).replace('\\','/')  # 原图片文件夹路径
                    print('oldFilePath', oldFilePath)
                    outputfilename.write(oldFilePath)
                    outputfilename.write('\r\n')

                    shutil.copy(oldFilePath, newFolderPath)

CombineFile()
outputfilename.close()
```
